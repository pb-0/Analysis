```python
import pandas as pd
```

```python
df = pd.read_csv('delta_price_symbol_X.csv', index_col=None)
```

```python
df_x = df.iloc[:, 1:]
```

```python
df = pd.read_csv('delta_price_symbols_E_to_H.csv')
df_EH = df.iloc[:, 1:]
```

```python
df = pd.read_csv('delta_price_symbols_A_to_D.csv')
df_AD = df.iloc[:, 1:]
```

```python
df_all = df_x.merge(df_EH, on='Date').merge(df_AD, on='Date')
```

```python
import statsmodels.api as sm
```

```python
y = df_all['delta_price_X']
```

```python
X = df_all.iloc[:,2:]
```

```python
results = sm.OLS(y, X).fit()

print(results.summary())
```

```python
infl_s = results.get_influence().summary_frame()
# df = infl['influence']
```

```python
infl_s.describe()
```

```python
fts = list(df_all.columns[2:])
```

```python
df_nrml = df_all
for x in fts:
    df_nrml[x] = pd.qcut(df_nrml[x], q=100, labels=False, duplicates='drop')
```
